```python
import torch
import librosa
from transformers import pipeline

# 1. Check if GPU is available and set the device
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# 2. Initialize the pipeline on the GPU
pipe = pipeline(
    task="automatic-speech-recognition",
    model="Lingalingeswaran/whisper-small-sinhala",
    device=device
)

# 3. Load your uploaded audio file (automatically resampled to 16kHz)
audio_path = "/content/omnivoice-audio-1781154173519.mp3"  # <-- Replace with your file name
audio, sampling_rate = librosa.load(audio_path, sr=16000)

# 4. Transcribe
result = pipe(
    audio,
    generate_kwargs={"task": "transcribe", "language": "si"}
)

# 5. Print the output
print("\n--- Transcription ---")
print(result["text"])
```

```python
from jiwer import wer, cer

# 1. Provide your Ground Truth (Manually verified Sinhala)
reference = "මම අයිස්ක්‍රීම් කන්න කැමතියි."

# 2. Provide the WhisperX output text
hypothesis = "මම අයිස්ක්‍රීම් කරන්න කැමති"

# Calculate Word Error Rate (WER)
error_rate = wer(reference, hypothesis)
accuracy = (1 - error_rate) * 100

# Calculate Character Error Rate (CER) - highly recommended for Sinhala script
char_error_rate = cer(reference, hypothesis)
char_accuracy = (1 - char_error_rate) * 100

print("--- Accuracy Metrics ---")
print(f"Word Error Rate (WER): {error_rate:.2%}")
print(f"Word Accuracy: {accuracy:.2f}%\n")

print(f"Character Error Rate (CER): {char_error_rate:.2%}")
print(f"Character Accuracy (Phonetic Precision): {char_accuracy:.2f}%")
```
